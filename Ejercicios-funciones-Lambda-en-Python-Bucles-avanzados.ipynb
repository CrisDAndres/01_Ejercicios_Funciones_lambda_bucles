# Ejercicio 3

Dada una lista y un número, cree una función que devuelva dos parámetros:

1) Si está dicho número en esa lista. En dicho caso Devolverá `TRUE`. En caso contrario, `False`.

2) Si la longitud de la lista coincide con dicho número. En dicho caso devolverá `TRUE`. En caso contrario, `False`.

````python
lista = [1, 2, 3, 4, 5]
def infor_lista (x):
    
    if x in lista:
        print('La lista contiene ese número')
        
        if x == len(lista):
            print('y además la longitud de la lista coincide con ese número')
        return True
    
    else:
        print('La lista no contiene ese número')
        return False
infor_lista (5)
````

````python
# Otra manera:
def doble_salida(lista,comprobar):
    presente = True if comprobar in lista else False
    longitud = True if len(lista) == comprobar else False 
    return presente,longitud

print(doble_salida([1,2,3,4,5], 5))
````

````python
# Usando lambda y listas de comprensión:

lista = [1, 2, 3, 4, 5]
numero = 5

programa = lambda lista, numero : (True if numero in lista else False, True if len(lista) == numero else  False) # Si queremos que lambda haga 2 cosas, separar con () y ,
print(programa(lista,numero))
````

# Ejercicio 4

Dada una lista, cree dos funciones, una que devuelva el mayor número y otra que devuelva en menor número.



````python
lista = [1, 11, 3, 4, 5]
def mayor (lista):
    lista.sort()
    
    return lista[-1]
def menor (lista):
    lista.sort()
    
    return lista[0]
    
print('El número mayor de la lista es {} y el menor es {}.'.format(mayor(lista), menor(lista)))
````

````python
# De otra manera:

lista = [1, 11, 3, 4, 5]
def mayor_menor (lista):

    return max(lista), min(lista)
    
mayor_menor(lista)
````

````python
# De una manera más optimizada:

lista = [1, 11, 3, 4, 5]
mayor_menor = lambda lista : (max(lista),min(lista)) # Si queremos que lambda haga 2 cosas, separar con () y ,
print('El número mayor y menor de la lista son', mayor_menor(lista))
print('El número mayor y menor de la lista {} son'.format(lista), mayor_menor(lista))
````

# Ejercicio 5
Use la funcion `sorted` para ordenar una lista de números en orden ascendente y también descendente.

# Ejercicio 7

Cree una función para contar la cantidad de números negativos de una lista.

````python
lista = [-1, 3, -7, 50]
def negativos(lista):
    negativos=0
    for numero in lista:
        if numero < 0:
            negativos+=1
    return negativos

negativos(lista)
````

````python
# Ejemplo compañero

lista = [-3,-5,7,9,2,45,-65,-4]

def contador_negativo(l):
    """
    Esta funcion indica la cantidad de numeros negativos en una lista
    """
    return sum(1 for num in l if num < 0)

contador_negativo(lista)
````

````python
# Otra manera:

lista = [-1, 3, -7, 50]
def negativos(lista):
    return [len([numero for numero in lista if numero < 0])] # La lista de comprensión genera una lista con los elementos que cumplan lo que digo, entonces hay que usar len para calcular la longitud

negativos(lista)
````

````python
# Utilizando lambda

negativos = lambda lista : len(list(filter(lambda x:x>0, lista)))
negativos(lista)
````

# Ejercicio 8

Cree una función para contar la cantidad de números impares de una lista.


````python
lista = [-1, 3, 2, 50,4,3]
def impares(lista):
    impares=0
    for x in lista:
        if x%2!=0:
            impares+=1
    return impares

impares(lista)
````

# Ejercicio de ampliación 1
Modifique el programa de la contraseña para dar al usuario solo 3 intentos. Si pasados dichos intentos el usuario no ha escrito bien la contraseña el programa acabará.

````python
intento = 0
password_right = 'Elefante'

while intento < 3: # Otra manera es for i in range(3)
    password = input('Inserte la contraseña. Dispone de 3 intentos.')
    if password != password_right:
        intento += 1
        print('Contraseña incorrecta. Tiene {} intentos restantes.'.format(3-intento))
    else:
        print("La contraseña es correcta.")
print('Ha superado el número de intentos. Acceso denegado')
````

````python
# Utilizando for y range en lugar de while
password_right = 'Elefante'

for i in range(3): # Otra manera es for i in range(3), y ya no necesito crear intentos = 0
    password = input('Inserte la contraseña. Dispone de 3 intentos.')
    if password == password_right:
        print('Contraseña correcta. Bienvenido!')
        break
    else:
        
        if i < 2: # Porque range va de 0 a 2
            print("La contraseña es incorrecta. Intentelo de nuevo")
        else:
            print('Ha superado el número de intentos. Acceso denegado')
````

**4**) Devuelva su matrix *traspuesta*.
[Matriz traspuesta](https://upload.wikimedia.org/wikipedia/commons/e/e4/Matrix_transpose.gif). Este punto puede realizarlo fuera de las condiciones realizadas anteriormente.

**Extra**) puede controlar si un usuario introduce otro valor (letra, símbolo u otro caracter) usando `isdigit`, `is_numeric` o `is_decimal`. ¿Cuál vendría mejor? 


**En todo momento trate de usar listas de comprensión, siempre que le sea posible.**

````python
matriz = [[1, 2], [3, 4], [5, 6], [7, 8]]
print('Filas:',len(matriz[0])) # Imprime las filas
print('Columnas:',len(matriz)) # Imprime las columnas
"""
Ahora vamos a generar la matriz traspuesta, cambiando las filas por las columnas y viceversa
"""
matriz_traspuesta = []
for columna in range(len(matriz[0])): # Recorremos cada columna elemento por elemento
    nueva_fila = []
    for fila in matriz: # Recorremos cada fila elemento por elemento
        nueva_fila.append(fila[columna]) # Agregamos los elementos a la nueva fila, cambiando las filas por columnas
    
    matriz_traspuesta.append(nueva_fila) # Se agrega la nueva fila a la matriz traspuesta

print('Matriz traspuesta:',matriz_traspuesta)
````
